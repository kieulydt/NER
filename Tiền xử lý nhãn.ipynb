```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./data/data_mobile_origin.csv')
```

```python
data.head()
```

```python
data.fillna(-1, inplace = True)
```

```python
data.head()
```

```python
# Xử lý nhãn hiệu
for i in range(len(data)):
    if data['nhan_hieu'][i] != -1:
        start_nhan_hieu = data['TEN_HANG'][i].find(data['nhan_hieu'][i])
        if start_nhan_hieu == -1:
            data['nhan_hieu'][i] = data['nhan_hieu'][i].replace(" ", "")
            
```

```python
for i in range(len(data)):
    if data['nhan_hieu'][i] != -1:
        start_nhan_hieu = data['TEN_HANG'][i].find(data['nhan_hieu'][i])
        if start_nhan_hieu == -1:
            print(i, data['nhan_hieu'][i])
```

```python
# Xử lý tình trạng
for i in range(len(data)):
    if data['tinh_trang'][i] != -1:
        start_so_hieu = data['TEN_HANG'][i].find(data['tinh_trang'][i])
        if start_so_hieu == -1:
            data['tinh_trang'][i] = 'mới 100%'
```

```python
count =0 
for i in range(len(data)):
    if data['tinh_trang'][i] != -1:
        start_so_hieu = data['TEN_HANG'][i].find(data['tinh_trang'][i])
        if start_so_hieu == -1:
            count +=1
            print(i, data['tinh_trang'][i])
print(count)
```

```python
data.to_csv('./data/data_revise_tag.csv')
```

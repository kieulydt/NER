```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./data/data_mobile_origin.csv')
```

```python
data.fillna(-1, inplace = True)
```

```python
data.head()
```

```python
import nltk
from nltk import ngrams
sentences = []
y_sentences = []

def ngram(text):
    return [i[0] for i in ngrams(text, n=1)]
for i in range(len(data)):
    # Chuyển chữ hoa thành chữ thường
    sentences.append(ngram(data['TEN_HANG'][i]))
    y_add = ['0'] * len(data['TEN_HANG'][i])
    
    # Tag nhãn hiệu
    if data['nhan_hieu'][i] != -1:
        start_nhan_hieu = data['TEN_HANG'][i].find(data['nhan_hieu'][i])
        if start_nhan_hieu != -1:
            y_add[start_nhan_hieu] = 'B_BRAND'
            for j in range(start_nhan_hieu+1, start_nhan_hieu + len(data['nhan_hieu'][i])):
               y_add[j] = 'I_BRAND'
    # Tag số hiệu
    if data['so_hieu'][i] != -1:
        start_so_hieu = data['TEN_HANG'][i].find(data['so_hieu'][i])
        if start_so_hieu != -1:
            y_add[start_so_hieu] = 'B_MODEL'
            for j in range(start_so_hieu+1, start_so_hieu+len(data['so_hieu'][i])):
                y_add[j] = 'I_MODEL'
    # Tag tình trạng
    if data['tinh_trang'][i] != -1:
        start_tinh_trang = data['TEN_HANG'][i].find(data['tinh_trang'][i])
        if start_tinh_trang != -1:
            y_add[start_tinh_trang] = 'B_STATUS'
            for j in range(start_tinh_trang+1, start_tinh_trang+len(data['tinh_trang'][i])):
                y_add[j] = 'I_STATUS'
    # Tag màu
    if data['color'][i] != -1:
        start_color = data['TEN_HANG'][i].find(data['color'][i])
        if start_color != -1:
            y_add[start_color] = 'B_COLOR'
            for j in range(start_color + 1, start_color + len(data['color'][i])):
                y_add[j] = 'I_COLOR'
    # Tag type
    if data['type'][i] != -1:
        start_type = data['TEN_HANG'][i].find(data['type'][i])
        if start_type != -1:
            y_add[start_type] = 'B_TYPE'
            for j in range(start_type +1, start_type+len(data['type'][i])):
                y_add[j] = 'I_TYPE'
    # Tag xuất xứ
    if data['xuat_xu'][i] != -1:
        start_xuat_xu = data['TEN_HANG'][i].find(data['xuat_xu'][i])
        if start_xuat_xu != -1:
            y_add[start_xuat_xu] = 'B_ORIGIN'
            for j in range(start_xuat_xu+1, start_xuat_xu+len(data['xuat_xu'][i])):
                y_add[j] = 'I_ORIGIN'
    y_sentences.append(y_add)
```

```python
all_chars = []
for sent in sentences:
    for item in sent:
        all_chars.append(item)
all_chars = set(all_chars)
all_chars = list(all_chars)
all_chars.sort()

all_tags = []
for y_sent in y_sentences:
    for item in y_sent:
        all_tags.append(item)
all_tags = set(all_tags)
all_tags = list(all_tags)
all_tags.sort()
```

```python
all_tags
```

```python
# Gán chỉ số cho các ký tự
char2ind = {value:ind+2 for ind, value in enumerate(all_chars)}
char2ind['UNK'] = 1
char2ind['PAD'] = 0

ind2char = {ind:value for value, ind in char2ind.items()}

# Gán chỉ số cho các nhãn
tag2ind = {value: ind+1 for ind, value in enumerate(all_tags)}
tag2ind['PAD'] = 0

ind2tag = {ind:value for value, ind in tag2ind.items()}
```

```python
MAX_LEN = max([len(sent) for sent in sentences])
```

```python
# Xây dựng tập X và Y để train
X = [[char2ind[item] for item in sent] for sent in sentences]
Y = [[tag2ind[tag] for tag in y_sent] for y_sent in y_sentences]
```

```python
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
```

```python
X_pad = pad_sequences(maxlen = MAX_LEN, sequences = X, padding = "post", value = char2ind['PAD'])
y_pad = pad_sequences(maxlen = MAX_LEN, sequences = Y, padding = "post", value = char2ind['PAD'])
y_categori = [to_categorical(ind, num_classes=len(all_tags)+1) for ind in y_pad]
```

```python
y_categori
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pad, y_categori, test_size = 0.3, random_state=0)
```

```python
X_train
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Embedding, Dense, Dropout, Bidirectional, TimeDistributed

model = Sequential()
model.add(Embedding(input_dim=len(all_chars) + 2, output_dim=100, input_length=MAX_LEN))
model.add(Dropout(0.5))
model.add(Bidirectional(LSTM(units=64, return_sequences=True, recurrent_dropout=0.1)))
model.add(Bidirectional(LSTM(units=128, return_sequences=True, recurrent_dropout=0.1)))
model.add(Dense(len(all_tags) + 1, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
#model.add(TimeDistributed(Dense(len(all_tags) + 1, activation="relu")))
#crf = CRF(len(all_tags) + 1, sparse_target=True)
#model.add(crf)
#model.compile(optimizer='adam', loss=crf.loss_function, metrics=[crf.accuracy])
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
```

```python
BATCH_SIZE = 128
EPOCHS = 20
checkpoint = ModelCheckpoint(
        './save/model.h5',
        monitor='val_loss',
        verbose=1,
        save_best_only=True)

history = model.fit(
        X_train,
        np.array(y_train),
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=[X_test, np.array(y_test)],
        verbose=1,
        callbacks=[checkpoint]
    )
```
